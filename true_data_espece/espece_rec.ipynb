```python
def add_abbreviations_to_dataframe(df, column_name, predefined_words, abbreviations):
    # cr√©er une liste vide pour stocker les abr√©viations
    column_values = []
    
    # boucle √† travers chaque ligne de la colonne et recherche les mots pr√©d√©finis
    for index, row in df.iterrows():
        content = row[column_name].lower()
        abbreviation = ''
        for i in range(len(predefined_words)):
            if predefined_words[i] in content:
                abbreviation += abbreviations[i]  # ajouter l'abr√©viation du mot cl√© correspondant
        column_values.append(abbreviation)
    
    # ajouter la colonne au DataFrame
    df['Abbreviations'] = column_values
    return df
```

```python
predefined_words = ['reg cheque','encaissement cheque', 'virement emis','virement recu','prelevement','telecompensation','versement esp','encaissement chq']
abbreviations = ['REG_CH', 'ENC_CH','VE','VR','PR','FR_TL','V_ESP','ENC_CH']
df = add_abbreviations_to_dataframe(data,'Libell√©', predefined_words, abbreviations)
```

```python
data1 = pd.read_csv("E://Stage_AIREC_2023//Extrait.csv", sep = ',')

print("Shape of data : " , data1.shape,"\n")

data1.head(14)
```

```python
#remplir les valeurs manquantes
df1=data1.fillna(0)
```

```python
# ajout d'une nvlle colonne contenant les id
df1.insert(0,'ID', df1.reset_index().index)
```

```python
df1
```

```python
# filtrer les lignes de df o√π la colonne 'abr√©viation' se termine par 'er'.
espece_df = df1[df1['Libell√©'].str.endswith('ES') | df1['Libell√©'].str.endswith('VB') ]
```

```python
espece_df
```

```python
# convertir les colonnes de date au format datetime
df['Date 1 '] = pd.to_datetime(df['Date 1 '])
espece_df['Date \nde r√©glement'] = pd.to_datetime(espece_df['Date \nde r√©glement'])
```

```python
#trier les valeurs de "Date 1 "
df_sorted = df.sort_values('Date 1 ')
```

```python
#trier les valeurs de "Date \nde r√©glement "
espece_df_sorted = espece_df.sort_values('Date \nde r√©glement')
```

```python
#fusionner les deux dataframe en tenant compte de date avec une tolerance de 2 jours d'ecart
merged_df = pd.merge_asof(df_sorted, espece_df_sorted, left_on='Date 1 ', right_on='Date \nde r√©glement', tolerance=pd.Timedelta('2 days'))
```

```python
merged_df
```

```python
from fuzzywuzzy import fuzz
# D√©finir une fonction pour calculer le taux de correspondance floue entre deux cha√Ænes de caract√®res.
def fuzzy_match(s1, s2):
    return fuzz.ratio(str(s1), str(s2))
```

```python
# Appliquez la fonction fuzzy_match aux colonnes "montant" et "cr√©dit" et obtenez les lignes o√π le ratio de correspondance est sup√©rieur √† 50.
matched_rows = merged_df[merged_df.apply(lambda row: fuzzy_match(row["montant"], row["Cr√©dit"]) > 50, axis=1)]
```
